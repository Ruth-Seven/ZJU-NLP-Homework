```python
import sys
import os 

from gensim.models.word2vec import Word2Vec 
from gensim.models.word2vec  import LineSentence
from gensim.test.utils import datapath 
```

```python
# sentence = LineSentence(datapath('zhwiki.text'))
output_path = 'zhwiki.text'
net = Word2Vec(LineSentence(output_path), size=400, window=5, min_count=5, workers=6)
net_output = 'wiki.zh.model'
net.save(net_output)
```

```python
net = Word2Vec.load(net_output)
# 华为词对应的numpy向量
net.wv['华为']

```
